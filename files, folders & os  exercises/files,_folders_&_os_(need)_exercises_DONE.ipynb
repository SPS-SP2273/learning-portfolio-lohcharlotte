```python
import os
import glob
import shutil

cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
          'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

parent_directory = "/Users/charmaineloh/Desktop/Cities"  

for city in cities:
    city_folder = os.path.join(parent_directory, city)
    os.makedirs(city_folder, exist_ok=True)

for city in cities:
    city_files = glob.glob(os.path.join(parent_directory, city, "*"))
    
    for file in city_files:
        shutil.move(file, os.path.join(parent_directory, city))
```
