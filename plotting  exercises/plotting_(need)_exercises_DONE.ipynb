```python
import matplotlib.pyplot as plt

# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = 0.25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='red', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black', capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=0.25)
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

x_values = np.arange(0, 51)

y_values = [math.exp(-x/10) * math.sin(x) for x in x_values]

plt.plot(x_values, y_values)
plt.xlabel('x', fontsize=15, color='darkorange')
plt.ylabel('y', fontsize=15, color='darkorange')
plt.title('$y=e^{-x/10}\sin(x)$', fontsize=20)
plt.grid(alpha=0.25)
plt.ylim([-1, 1])
plt.show()
```
