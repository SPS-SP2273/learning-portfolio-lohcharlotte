```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
```

```python
# Plotting the cosine function 
plt.plot(x, cos_x, label='cos x')
# Plotting the sine function 
plt.plot(x, sin_x, label='sin x')
plt.legend() #Add legend 
fig, ax = plt.subplots(nrows=1, ncols=1) #Creating a figure and one subplot
ax.plot(x, cos_x, label='cos x') #Plotting cosine values against x on subplot
ax.plot(x, sin_x, label='sin x') #Plotting sine values against x on subplot
ax.legend()
```

```python
# Plotting the cosine function 
plt.plot(x, -cos_x, label='-cos x')
# Plotting the sine function 
plt.plot(x, -sin_x, label='-sin x')
plt.legend() #Add legend 
fig, ax = plt.subplots(nrows=1, ncols=1) #Creating a figure and one subplot
ax.plot(x, -cos_x, label='-cos x') #Plotting cosine values against x on subplot
ax.plot(x, -sin_x, label='-sin x') #Plotting sine values against x on subplot
ax.legend()
```

# 4 A complete OO example

```python
# Creating a figure with two subplots vertically stacked,
# setting the figure size to (5, 5), and sharing the x-axis
fig, ax = plt.subplots(nrows=2, ncols=1,
                       figsize=(5, 5),
                       sharex=True)
# Plotting the cosine function on the first subplot
ax[0].plot(x, cos_x, label=r'$\cos(x)$')
# Filling the area under the cosine curve with transparency
ax[0].fill_between(x, 0, cos_x, alpha=.25)
# Plotting the sine function on the second subplot
ax[1].plot(x, sin_x, label=r'$\sin(x)$')
# Filling the area under the sine curve with transparency
ax[1].fill_between(x, 0, sin_x, alpha=.25)

# Adding legend to both 
for a in ax:
    a.legend()
    # Adding grid lines with transparency
    a.grid(alpha=.25)
    # Setting y-axis label
    a.set_ylabel('$y$', rotation=0)

# Setting x-axis label for the second subplot
ax[1].set_xlabel('$x$')

# Adding a title to the entire figure
fig.suptitle(r'$\sin(x)$ and $\cos(x)$')
# Adjusting layout to prevent overlapping elements
fig.tight_layout()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1,
                       figsize=(5, 5),
                       sharex=True)
ax[0].plot(x, cos_x, label=r'$\cos(x)$')
ax[0].fill_between(x, 0, cos_x, alpha=.25)
ax[1].plot(x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25)

for a in ax:
    a.legend()
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0)

ax[1].set_xlabel('$x$')

fig.suptitle(r'$\sin(x)$ and $\cos(x)$')
fig.tight_layout()
```

# 5 Multiple rows and columns

```python
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(5, 5),
                       sharex='col', sharey='col')

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```

## 6.1 Histograms

### A quick helper function

```python
def get_plot_data():
    data = {}
    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_all.csv'
    data['All'] = np.loadtxt(filename, skiprows=1)

    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_nus.csv'
    data['NUS'] = np.loadtxt(filename, skiprows=1)

    return data
```

### The histogram

```python
plt.style.use('bmh')
data = get_plot_data()

# bins specifies how many bins to split the data
plt.hist([data['All'], data['NUS']], bins=50, label=['All', 'NUS'])
plt.xlabel('Mean of Basic Montly Earning (S$)')
plt.ylabel('Number of Students')
plt.legend()
```

## 6.3 Bar charts

```python
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
majors = list(student_numbers.keys())
numbers = list(student_numbers.values())

plt.style.use('ggplot')
plt.bar(majors, numbers)
plt.xlabel('Majors')
plt.ylabel('Number of Students')
```

## 6.4 Pie charts

```python
# Dictionary containing the number of students for each major
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
# Extracting majors and corresponding student numbers
majors = list(student_numbers.keys()) # Extracting majors
numbers = list(student_numbers.values()) # Extracting student numbers

# Setting the style for the plot
plt.style.use('fivethirtyeight')

# Creating a pie chart
plt.pie(numbers, 
        labels=majors,
        autopct='%1.1f%%',   # How to format the percentages
        startangle=-90                
        )

# Adding a title to the plot
plt.title('Percentage of each major')
```
